## 2. Pemilihan baris (rows selection) pada `DataFrame`

## Import modules

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 5
cols = list("ABCDE")

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), columns=cols)
df
```

## Selection dengan operator logika `|`

```python
df[(df["A"] == 1) | (df["A"] == 3)]
```

## Selection dengan method `.isin()`

```python
df[df["A"].isin([1, 3])]
```

## Selection dengan operator `~`

```python
df[~df["A"].isin([1, 3])]
```
